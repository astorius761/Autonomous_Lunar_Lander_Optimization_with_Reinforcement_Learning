```python
import torch
import numpy as np
import gymnasium as gym
import cv2
import matplotlib.pyplot as plt
import pickle
import time
from collections import deque


try:
    import google.colab
    IN_COLAB = True
    from google.colab import drive
    from google.colab.patches import cv2_imshow
except ImportError:
    IN_COLAB = False

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

class LunarLanderQAgent:
    def __init__(self, grid_size=4, learning_rate=0.1,
                 gamma=0.99, epsilon_start=1.0, epsilon_min=0.01):

        self.env = gym.make("LunarLander-v3", render_mode="rgb_array")
        self.action_space = self.env.action_space.n


        self.q_table = {}
        self.alpha = learning_rate
        self.gamma = gamma
        self.epsilon = epsilon_start
        self.epsilon_min = epsilon_min


        self.epsilon_decay_amount = 0.001
        self.epsilon_decay_frequency = 10


        self.full_exploration_episodes = 1300  #1000
        self.current_episode = 0



        self.grid_size = grid_size
        self.target_size = (84, 84) # (16, 16)    (32, 32)




        self.rewards_history = []
        self.epsilon_history = []
        self.q_table_size_history = []

        # landing
        self.prev_features = None
        self.prev_distance_to_pad = None
        self.step_count = 0
        self.total_fuel_used = 0
        self.start_height = None

        # Previous state velocities
        self._prev_center_x = 0.5
        self._prev_center_y = 0.5

    def preprocess_image(self, image):


        if isinstance(image, torch.Tensor):
            image = image.cpu().numpy()


        if len(image.shape) == 3 and image.shape[2] == 3:
            gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
        else:
            gray = image


        resized = cv2.resize(gray, self.target_size)



        normalized = resized / 255.0

        #to torch
        tensor = torch.tensor(normalized, dtype=torch.float32, device=device)

        return tensor

    def extract_features(self, processed_image):

        if isinstance(processed_image, torch.Tensor):
            processed_image = processed_image.cpu().numpy()

        h, w = processed_image.shape
        grid_features = []


        for i in range(self.grid_size):
            for j in range(self.grid_size):
                region = processed_image[i*h//self.grid_size:(i+1)*h//self.grid_size,
                                      j*w//self.grid_size:(j+1)*w//self.grid_size]

                grid_features.append(1 if np.mean(region) > 0.2 else 0)


        landing_pad_region = processed_image[h//2:, :]
        pad_x = np.argmax(np.sum(landing_pad_region > 0.5, axis=0))
        pad_x_bin = min(int(pad_x / w * 5), 4)


        lander_region = processed_image[:h//2, :]
        if np.max(lander_region) > 0.3:  # If lander is visible
            nonzero_y, nonzero_x = np.where(lander_region > 0.3)
            if len(nonzero_y) > 0 and len(nonzero_x) > 0:
                lander_x = np.mean(nonzero_x)
                lander_y = np.mean(nonzero_y)
                lander_x_bin = min(int(lander_x / w * 5), 4)  #
                lander_y_bin = min(int(lander_y / (h//2) * 5), 4)
            else:
                lander_x_bin, lander_y_bin = 0, 0
        else:
            lander_x_bin, lander_y_bin = 0, 0


        features = tuple(grid_features + [lander_x_bin, lander_y_bin, pad_x_bin])
        return features

    def estimate_lander_state(self, processed_image):

        if isinstance(processed_image, torch.Tensor):
            processed_image = processed_image.cpu().numpy()

        h, w = processed_image.shape


        binary = (processed_image > 0.3).astype(np.uint8)


        lander_region = binary[:h//2, :]#up
        lander_y, lander_x = np.where(lander_region > 0)

        pad_region = binary[h//2:, :]#down
        pad_x = np.argmax(np.sum(pad_region, axis=0)) if np.sum(pad_region) > 0 else w//2

        if len(lander_y) > 0 and len(lander_x) > 0:

            lander_center_x = np.mean(lander_x) / w
            lander_center_y = np.mean(lander_y) / (h//2)


            distance_to_pad = abs(lander_center_x - (pad_x / w))

            left_side = np.sum(lander_region[:, :int(lander_center_x * w)])#center
            right_side = np.sum(lander_region[:, int(lander_center_x * w):])
            tilt = (right_side - left_side) / max(1, left_side + right_side)

            height = 1 - lander_center_y

            vel_x = lander_center_x - self._prev_center_x
            vel_y = lander_center_y - self._prev_center_y

            # updqt
            self._prev_center_x = lander_center_x
            self._prev_center_y = lander_center_y

            return {
                'distance_to_pad': distance_to_pad,
                'height': height,
                'tilt': tilt,
                'lander_x': lander_center_x,
                'pad_x': pad_x / w,
                'vel_x': vel_x,
                'vel_y': vel_y,
                'visible': True
            }
        else:

          #ajusting if the landing failed
            return {
                'distance_to_pad': 1.0,
                'height': 1.0,
                'tilt': 0.0,
                'lander_x': 0.5,
                'pad_x': pad_x / w,
                'vel_x': 0,
                'vel_y': 0,
                'visible': False
            }

    def custom_reward(self, original_reward, lander_state, action, done):

        modified_reward = original_reward


        distance_to_pad = lander_state['distance_to_pad']
        height = lander_state['height']
        tilt = lander_state['tilt']
        visible = lander_state['visible']
        vel_x = lander_state['vel_x']
        vel_y = lander_state['vel_y']


        #reward for getting closer to the landing pad
        if self.prev_distance_to_pad is not None and visible:
            distance_improvement = self.prev_distance_to_pad - distance_to_pad
            modified_reward += distance_improvement * 10  # Reward for horizontal alignment

        #penalize excessive tilting/rotation
        tilt_penalty = -abs(tilt) * 5
        modified_reward += tilt_penalty

        #Reward fuel efficiency penalize engine usage


        # Main engine (action 2) uses more fuel than side engines (actions 1 and 3)

        if action == 2:
            fuel_penalty = -0.05
        elif action in [1, 3]:
            fuel_penalty = -0.02
        else:
            fuel_penalty = 0.01
        modified_reward += fuel_penalty

        # 4. Time-based penalty to encourage faster landing
        time_penalty = -0.01 * self.step_count
        modified_reward += time_penalty

        # 5. improve gardiant for smooth landing + slow decent
        if self.start_height is not None and visible:
            ideal_descent_rate = self.start_height / 200
            current_descent_rate = (self.start_height - height) / max(1, self.step_count)
            descent_rate_diff = abs(current_descent_rate - ideal_descent_rate)
            descent_reward = 0.1 * (1 - min(1, descent_rate_diff * 10))
            modified_reward += descent_reward

        # 6.penalize high horizontal velocity
        if visible:
            horiz_vel_penalty = -abs(vel_x) * 5
            modified_reward += horiz_vel_penalty

        # 7.extra reward for successful landing
        if done and original_reward > 0:
            modified_reward += 50  # Bonus for successful completion

        # 8.Cap extreme negative rewards to prevent training instability
        modified_reward = max(modified_reward, -10)

        #Updating
        self.prev_distance_to_pad = distance_to_pad

        return modified_reward

    def choose_action(self, state_features):

        if self.current_episode < self.full_exploration_episodes:
            return np.random.randint(self.action_space)

        # Exploration
        if np.random.random() < self.epsilon:
            return np.random.randint(self.action_space)

        # Exploitation: choose best known action
        if state_features in self.q_table:
            q_values = torch.tensor(self.q_table[state_features], device=device)
            return torch.argmax(q_values).item()
        else:
            #initialize q values for this state
            self.q_table[state_features] = torch.zeros(self.action_space, device=device)
            return np.random.randint(self.action_space)

    def update_q_table(self, state_features, action, reward, next_state_features, done):

        #ini unkown stat
        if state_features not in self.q_table:
            self.q_table[state_features] = torch.zeros(self.action_space, device=device)

        if next_state_features not in self.q_table:
            self.q_table[next_state_features] = torch.zeros(self.action_space, device=device)

        #update
        reward_tensor = torch.tensor(reward, device=device)
        done_tensor = torch.tensor(1.0 if done else 0.0, device=device)

        current_q = self.q_table[state_features][action]

        next_q_max = torch.max(self.q_table[next_state_features])

        td_target = reward_tensor + self.gamma * next_q_max * (1 - done_tensor)# current action + the best predict action i can get

        td_error = td_target - current_q #how far the predict was for improvment

        # Update Q-value
        self.q_table[state_features][action] += self.alpha * td_error

    def decay_epsilon(self):
        if self.current_episode >= self.full_exploration_episodes:
            if (self.current_episode - self.full_exploration_episodes) % self.epsilon_decay_frequency == 0:
                self.epsilon = max(self.epsilon_min, self.epsilon - self.epsilon_decay_amount)

    def train(self, n_episodes=2000, max_steps=1000, save_freq=100):

        save_path = "./"


        if IN_COLAB:
            try:
                drive.mount('/content/drive')
                save_path = "/content/drive/MyDrive/lunar_lander_models/"

                import os
                os.makedirs(save_path, exist_ok=True)
                print(f"Google Drive mounted. Saving to {save_path}")
            except:
                print("Error mounting Google Drive. Saving locally instead.")


        for episode in range(1, n_episodes+1):
            self.current_episode = episode


            state, info = self.env.reset()

            frame = self.env.render()
            processed_state = self.preprocess_image(frame)
            state_features = self.extract_features(processed_state)
            lander_state = self.estimate_lander_state(processed_state)

            #ep trackers
            self.prev_features = state_features
            self.prev_distance_to_pad = lander_state['distance_to_pad']
            self.step_count = 0
            self.total_fuel_used = 0
            self.start_height = lander_state['height']
            self._prev_center_x = 0.5
            self._prev_center_y = 0.5
            #ep var ini
            total_reward = 0
            done = False

            #tarining
            for step in range(max_steps):
                self.step_count = step
                action = self.choose_action(state_features)
                next_state, reward, terminated, truncated, info = self.env.step(action)
                done = terminated or truncated
                frame = self.env.render()
                processed_next_state = self.preprocess_image(frame)
                next_state_features = self.extract_features(processed_next_state)
                next_lander_state = self.estimate_lander_state(processed_next_state)
                modified_reward = self.custom_reward(reward, next_lander_state, action, done)
                self.update_q_table(state_features, action, modified_reward, next_state_features, done)

                state = next_state
                processed_state = processed_next_state
                state_features = next_state_features
                total_reward += reward

                if done:
                    break

            #save
            self.rewards_history.append(total_reward)
            self.epsilon_history.append(self.epsilon)
            self.q_table_size_history.append(len(self.q_table))
            self.decay_epsilon()

            if episode % 10 == 0:
                avg_reward = np.mean(self.rewards_history[-10:])
                print(f"Episode: {episode}, | Avg Reward: {avg_reward:.2f} | Epsilon: {self.epsilon:.4f} | Q-table size: {len(self.q_table)}")

            if episode % save_freq == 0:
                self.save_model(f"{save_path}lunar_lander_qtable_ep{episode}.pkl")

        # PLTT
        self.plot_rewards()

        #traingn vid
        self.display_last_training_episode()

    def display_last_training_episode(self):

        last_episode_env = gym.make("LunarLander-v3", render_mode="rgb_array")
        state, info = last_episode_env.reset()


        frame = last_episode_env.render()
        processed_state = self.preprocess_image(frame)
        state_features = self.extract_features(processed_state)

        frames = []
        total_reward = 0
        done = False

        print("\nDisplaying last training episode:")

        while not done:

            frame = last_episode_env.render()
            frames.append(frame)

            if state_features in self.q_table:
                action = torch.argmax(self.q_table[state_features]).item()
            else:
                action = 0

            next_state, reward, terminated, truncated, info = last_episode_env.step(action)
            done = terminated or truncated

            next_frame = last_episode_env.render()
            processed_next_state = self.preprocess_image(next_frame)
            next_state_features = self.extract_features(processed_next_state)


            state = next_state
            state_features = next_state_features
            total_reward += reward

        print(f"Last training episode reward: {total_reward:.2f}")


        if IN_COLAB:
            self.display_frames_as_gif(frames)

    def save_model(self, filename):
        numpy_q_table = {}
        for key, value in self.q_table.items():
            numpy_q_table[key] = value.cpu().numpy()

        with open(filename, 'wb') as f:
            pickle.dump(numpy_q_table, f)
        print(f"Model saved to {filename}")

    def load_model(self, filename):

        with open(filename, 'rb') as f:
            numpy_q_table = pickle.load(f)
        self.q_table = {}
        for key, value in numpy_q_table.items():
            self.q_table[key] = torch.tensor(value, device=device)

        print(f"Model loaded from {filename}")

    def plot_rewards(self):
        plt.figure(figsize=(10, 6))
        rewards_np = [r.item() if isinstance(r, torch.Tensor) else r for r in self.rewards_history]
        plt.plot(rewards_np, alpha=0.4, label='Raw Rewards')

        if len(rewards_np) > 10:
            smoothed_rewards = []
            window_size = 10
            for i in range(window_size, len(rewards_np)):
                smoothed_rewards.append(np.mean(rewards_np[i-window_size:i]))
            plt.plot(range(window_size, len(rewards_np)), smoothed_rewards, 'r-',
                     linewidth=2, label='Smoothed Rewards (10-episode window)')


        if self.epsilon_history:
            plt.figure(figsize=(10, 6))
            plt.plot(self.epsilon_history, 'g-', label='Epsilon Value')
            plt.title('Epsilon Decay during Training')
            plt.xlabel('Episode')
            plt.ylabel('Epsilon')
            plt.grid(True, alpha=0.3)
            plt.savefig(f"epsilon_decay.png")
            plt.close()

        plt.title('Training Rewards per Episode')
        plt.xlabel('Episode')
        plt.ylabel('Total Reward')
        plt.legend()
        plt.grid(True, alpha=0.3)


        plt.tight_layout()
        plt.savefig(f"training_rewards.png")
        plt.show()
        plt.close()

    def evaluate(self, n_episodes=10):

        eval_env = gym.make("LunarLander-v3", render_mode="rgb_array")

        total_rewards = []

        for episode in range(n_episodes):
            state, info = eval_env.reset()

            frame = eval_env.render()
            processed_state = self.preprocess_image(frame)
            state_features = self.extract_features(processed_state)

            episode_reward = 0
            done = False
            frames = []

            while not done:

                frame = eval_env.render()
                frames.append(frame)
                if state_features in self.q_table:
                    action = torch.argmax(self.q_table[state_features]).item()
                else:
                    action = 0
                next_state, reward, terminated, truncated, info = eval_env.step(action)
                done = terminated or truncated

                next_frame = eval_env.render()
                processed_next_state = self.preprocess_image(next_frame)
                next_state_features = self.extract_features(processed_next_state)

                state = next_state
                state_features = next_state_features
                episode_reward += reward

            total_rewards.append(episode_reward)
            print(f"Evaluation Episode {episode+1}/{n_episodes}, Reward: {episode_reward:.2f}")

            if episode == n_episodes - 1 and IN_COLAB:
                self.display_frames_as_gif(frames)

        avg_reward = np.mean(total_rewards)
        print(f"\nEvaluation complete. Average reward: {avg_reward:.2f}")
        plt.close()
        return avg_reward

    def display_frames_as_gif(self, frames):

        try:
            from IPython.display import display, HTML
            import matplotlib.animation as animation

            plt.figure(figsize=(frames[0].shape[1] / 72.0, frames[0].shape[0] / 72.0), dpi=72)
            patch = plt.imshow(frames[0])
            plt.axis('off')

            def animate(i):
                patch.set_data(frames[i])

            anim = animation.FuncAnimation(plt.gcf(), animate, frames=len(frames), interval=50)
            display(HTML(anim.to_jshtml()))
            plt.close()
        except:
            print("Unable to display animation in this environment")


if __name__ == "__main__":

    agent = LunarLanderQAgent(
        grid_size=4,
        learning_rate=0.0068, #0001   0.0016   0.2  1-4e   1-2e   0.1
        gamma=0.99,              #1
        epsilon_start=1.0,
        epsilon_min=0.01
    )

    agent.train(n_episodes=10000, max_steps=1000, save_freq=100)
    agent.save_model("lunar_lander_final_model.pkl")
    agent.evaluate(n_episodes=5)
```
